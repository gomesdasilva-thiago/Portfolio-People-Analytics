```python
# Carregando pacotes
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
# Carregando base de dados
data = pd.read_excel('base.xlsx')
```

```python
# Visualizando as primeiras linhas
data.head()
```

```python
data.columns
```

```python
# Verificando os dados
data.describe()
```

```python
# Verificando registros por grupo
print('Máximo de registros por grupo', data['Group'].value_counts().max())
print('Mínimo de registros por grupo', data['Group'].value_counts().min())
```

```python
# Verificando NA's
pd.isna(data).sum()
```

```python
# Excluindo linhas inteiras vazias
data_1 = data.dropna(subset=['Participation'])
data_1.describe()
```

```python
# # Verificando NA's após exclusão das linhas vazias
conta_nulos = pd.isna(data_1).sum().to_frame().reset_index().rename(columns = {'index': 'coluna', 0: 'total nulos'})
conta_nulos
```

```python
# Criando dataframe com total de linhas por coluna
conta_linhas = data_1.count().to_frame().reset_index().rename(columns = {'index': 'coluna', 0: 'total linhas'})
conta_linhas
```

```python
# Juntando contagem de nulos por colunas com total de linhas por coluna e % de nulos por coluna
# Percentual de nulos por coluna baixo, então iremos remover as linhas com valor nulo em qualquer coluna

df_nulos_linhas = pd.merge(conta_nulos, conta_linhas, how = "inner", on = 'coluna')
df_nulos_linhas['% nulo'] = (df_nulos_linhas['total nulos'] / df_nulos_linhas['total linhas'])*100

df_nulos_linhas
```

```python
# Removendo linhas com
data_2 = data_1.dropna()
```

```python
data_2.describe()
```

```python
data_2.head()
```

```python
# Análise de correlação
correlation = data_2.corr()
```

```python
# Plot da matriz de correlação
fig, ax = plt.subplots(figsize=(20,20))
sn.heatmap(correlation, vmin = 0, vmax=1, center = 0)
```
