```python
# Verificando NA's
pd.isna(data).sum()
```

```python
# Excluindo linhas inteiras vazias
data_1 = data.dropna(subset=['Participation'])
data_1.describe()
```

```python
# # Verificando NA's apÃ³s exclusÃ£o das linhas vazias
conta_nulos = pd.isna(data_1).sum().to_frame().reset_index().rename(columns = {'index': 'coluna', 0: 'total nulos'})
conta_nulos
```

```python
# Criando dataframe com total de linhas por coluna
conta_linhas = data_1.count().to_frame().reset_index().rename(columns = {'index': 'coluna', 0: 'total linhas'})
conta_linhas
```

```python
# Juntando contagem de nulos por colunas com total de linhas por coluna e % de nulos por coluna
# Percentual de nulos por coluna baixo, entÃ£o iremos remover as linhas com valor nulo em qualquer coluna

df_nulos_linhas = pd.merge(conta_nulos, conta_linhas, how = "inner", on = 'coluna')
df_nulos_linhas['% nulo'] = (df_nulos_linhas['total nulos'] / df_nulos_linhas['total linhas'])*100

df_nulos_linhas
```

```python
# Removendo linhas com valores nulos
data_2 = pd.DataFrame(data_1.dropna().reset_index(drop = True))
```
